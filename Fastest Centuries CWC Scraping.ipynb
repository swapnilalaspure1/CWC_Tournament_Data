```python
# Importing Required Library
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
# Set the path to the ChromeDriver executable
chrome_driver_path = r"C:\Users\Swapnil Alaspure\Documents\MIT Exam 2nd\WebScraping\chromedriver.exe"

# Create ChromeOptions object
chrome_options = webdriver.ChromeOptions()

# Instantiate the Chrome WebDriver with the specified options
driver = webdriver.Chrome(options=chrome_options)

# Open the URL
driver.get("https://www.cricketworldcup.com/stats/players/fastest-hundreds")
```

```python
# Our list empty list
table_headers = []

name_xpath = '//*[@id="main-content"]/div/div/div[2]/div/div/table/thead/tr'
name = driver.find_element(By.XPATH, name_xpath).text
table_headers.append(name.split())


# Display the list
table_headers
```

```python
# Our list empty list
players_data = []

for i in range(1, 26):
    name_xpath = '//*[@id="main-content"]/div/div/div[2]/div/div/table/tbody/tr[' + str(i) + ']'
    name = driver.find_element(By.XPATH, name_xpath).text
    players_data.append(name.replace(' ', ','))


# Display the list
players_data
```

```python
# Positions

Pos = []

for i in range(1,26):
    
    pos_xpath='//*[@id="main-content"]/div/div/div[2]/div/div/table/tbody/tr['+ str(i) +']/td[4]'
    
    pos = driver.find_element(By.XPATH,
                                 pos_xpath).text
    
    Pos.append(pos)
    
Pos
```

```python
# Name

Name = []

for i in range(1,26):
    
    name_xpath='//*[@id="main-content"]/div/div/div[2]/div/div/table/tbody/tr[' + str(i) + ']/td[5]'
    
    name = driver.find_element(By.XPATH,
                                 name_xpath).text
    
    Name.append(name)
    
Name
```

```python
# Country

Country = []

for i in range(1,26):
    
    country_xpath='//*[@id="main-content"]/div/div/div[2]/div/div/table/tbody/tr[' + str(i) + ']/td[6]/span[2]'
    
    country = driver.find_element(By.XPATH,
                                   country_xpath).text
    
    Country.append(country)
    
Country
```

```python
# Deliveries

Deliveries = []

for i in range(1,26):
    
    Deliveries_xpath='//*[@id="main-content"]/div/div/div[2]/div/div/table/tbody/tr[' + str(i) + ']/td[7]/span'
    
    deliveries = driver.find_element(By.XPATH,
                                        Deliveries_xpath).text
    
    Deliveries.append(deliveries)
    
Deliveries
```

```python
# Against

Against = []

for i in range(1,26):
    
    against_xpath='//*[@id="main-content"]/div/div/div[2]/div/div/table/tbody/tr[' + str(i) + ']/td[10]'
    
    against = driver.find_element(By.XPATH,
                                   against_xpath).text
    
    Against.append(against)
    
Against
```

```python
# Venue

Venue = []

for i in range(1,26):
    
    Venue_xpath='//*[@id="main-content"]/div/div/div[2]/div/div/table/tbody/tr[' + str(i) + ']/td[11]'
      
    venue = driver.find_element(By.XPATH,
                                 Venue_xpath).text
    
    Venue.append(venue)
    
Venue
```

```python
# Match Date

MatchDate = []

for i in range(1,26):
    
    MatchDate_xpath='//*[@id="main-content"]/div/div/div[2]/div/div/table/tbody/tr[' + str(i) + ']/td[12]'
    
    match_date = driver.find_element(By.XPATH,
                                  MatchDate_xpath).text
    
    MatchDate.append(match_date)
    
MatchDate
```

```python
# To Close the driver
driver.close()
```

```python
# Creating Data Frame
Fastest_Centuries = pd.DataFrame({'Pos':Pos,
                             "Name":Name,
                             "Country":Country,
                             "Deliveries":Deliveries,
                             "Against":Against,
                             "Venue":Venue,
                             "Match Date":MatchDate,
                            })
Fastest_Centuries
```
