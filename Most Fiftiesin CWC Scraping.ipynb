```python
# Importing Required Library
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
# Set the path to the ChromeDriver executable
chrome_driver_path = r"C:\Users\Swapnil Alaspure\Documents\MIT Exam 2nd\WebScraping\chromedriver.exe"

# Create ChromeOptions object
chrome_options = webdriver.ChromeOptions()

# Instantiate the Chrome WebDriver with the specified options
driver = webdriver.Chrome(options=chrome_options)

# Open the URL
driver.get("https://www.cricketworldcup.com/stats/players/most-fifties")
```

```python
# Our list empty list
table_headers = []

name_xpath = '//*[@id="main-content"]/div/div/div[2]/div/div/table/thead/tr'
name = driver.find_element(By.XPATH, name_xpath).text
table_headers.append(name.split())


# Display the list
table_headers
```

```python
# Our list empty list
players_data = []

for i in range(1, 26):
    name_xpath = '//*[@id="main-content"]/div/div/div[2]/div/div/table/tbody/tr[' + str(i) + ']'
    name = driver.find_element(By.XPATH, name_xpath).text
    players_data.append(name.replace(' ', ','))


# Display the list
players_data
```

```python
# Positions

Pos = []

for i in range(1,26):
    
    pos_xpath='//*[@id="main-content"]/div/div/div[2]/div/div/table/tbody/tr['+ str(i) +']/td[4]'
    
    pos = driver.find_element(By.XPATH,
                                 pos_xpath).text
    
    Pos.append(pos)
    
Pos
```

```python
# Name

Name = []

for i in range(1,26):
    
    name_xpath='//*[@id="main-content"]/div/div/div[2]/div/div/table/tbody/tr[' + str(i) + ']/td[5]'
    
    name = driver.find_element(By.XPATH,
                                 name_xpath).text
    
    Name.append(name)
    
Name
```

```python
# Country

Country = []

for i in range(1,26):
    
    country_xpath='//*[@id="main-content"]/div/div/div[2]/div/div/table/tbody/tr[' + str(i) + ']/td[6]/span[2]'
    
    country = driver.find_element(By.XPATH,
                                 country_xpath).text
    
    Country.append(country)
    
Country
```

```python
# Hs Runs

Runs = []

for i in range(1,26):
    
    runs_xpath='//*[@id="main-content"]/div/div/div[2]/div/div/table/tbody/tr[' + str(i) + ']/td[10]'
    
    run = driver.find_element(By.XPATH,
                                 runs_xpath).text
    
    Runs.append(run)
    
Runs
```

```python
# Innings

Inn = []

for i in range(1,26):
    
    inn_xpath='//*[@id="main-content"]/div/div/div[2]/div/div/table/tbody/tr[' + str(i) + ']/td[8]'
    
    inn = driver.find_element(By.XPATH,
                                 inn_xpath).text
    
    Inn.append(inn)
    
Inn
```

```python
# Fifties 50s

fifties = []

for i in range(1,26):
    
    fifties_xpath='//*[@id="main-content"]/div/div/div[2]/div/div/table/tbody/tr[' + str(i) + ']/td[7]/span'
    
    Fifties = driver.find_element(By.XPATH,
                                 fifties_xpath).text
    
    fifties.append(Fifties)
    
fifties
```

```python
# Batting Average

Bat_Avg = []

for i in range(1,26):
    
    batAvg_xpath='//*[@id="main-content"]/div/div/div[2]/div/div/table/tbody/tr[' + str(i) + ']/td[9]'
    
    batAvg = driver.find_element(By.XPATH,
                                 batAvg_xpath).text
    
    Bat_Avg.append(batAvg)
    
Bat_Avg
```

```python
# To Close the driver
driver.close()
```

```python
# Creating Data Frame
Most_Fifties = pd.DataFrame({'Pos':Pos,
                             "Name":Name,
                             "Country":Country,
                             "Inn":Inn,
                             "50s":fifties,
                             "Bat Avg":Bat_Avg,
                             "Runs":Runs,
                            })
Most_Fifties
```
